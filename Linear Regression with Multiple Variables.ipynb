```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('car data.csv')
df.head()
```

# Mean
#  It is the average of all the numbers in a dataset.


```python
# there's no value in the 4th row of experience column. 
# It will be replaced with the mean value of the values of the column
df.experience.mean()
```

# Or, Median
 # It is the middle value of a dataset when the numbers are arranged in order (smallest to largest).


```python
# there's no value in the 4th row of experience column. 
# It will be replaced with the median value of the values of the column
df.experience.median()
```

# We'll go with the Median

```python
exp_fit = df.experience.median()
exp_fit
```

```python
# to fill null values, fillna is used
df.experience = df.experience.fillna(exp_fit)
df
```

```python
reg = linear_model.LinearRegression()
```

# Training the model: 

```python
# for multiple variables: 
reg.fit(df[['speed','car_age','experience']],df.risk)
```

```python
# predict for 160 speed, 10 years age, 5 years experience: 
reg.predict([[160,10,5]])
```
