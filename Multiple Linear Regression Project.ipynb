# Scatter Plot: Marketing Spend vs Profit

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Marketing Spend'], df['Profit'], color='blue', alpha=0.6)
plt.title('Marketing Spend vs Profit')
plt.xlabel('Marketing Spend')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

 # Bar Plot: Total Profit by Area

```python
profit_by_area = df.groupby('Area')['Profit'].sum()

plt.figure(figsize=(8, 5))
profit_by_area.plot(kind='bar', color=['red', 'green', 'blue'], edgecolor='black')
plt.title('Total Profit by Area')
plt.xlabel('Area')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```

 # Regression Line Plot: Actual vs Predicted Profit

```python
plt.figure(figsize=(10, 6))
plt.scatter(ytest, pred, color='purple', alpha=0.6, label='Predicted')
plt.plot([min(ytest), max(ytest)], [min(ytest), max(ytest)], 'k--', lw=2, label='Perfect Fit')
plt.title('Actual vs Predicted Profit')
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.legend()
plt.grid(True)
plt.show()
```

# Residual Plot: Assessing Model Fit

```python
residuals = ytest - pred

plt.figure(figsize=(10, 6))
plt.scatter(pred, residuals, color='orange', alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Residual Plot')
plt.xlabel('Predicted Profit')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()
```
